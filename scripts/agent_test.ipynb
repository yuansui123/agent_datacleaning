```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

# LOAD .env (should contain: OPENAI_API_KEY=sk-...)
load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY")
if not openai_api_key:
    raise RuntimeError("OPENAI_API_KEY not set. Add it to .env or export it before running.")

# Initialize model client
llm = ChatOpenAI(
    model="gpt-4",
    temperature=0,
    openai_api_key=openai_api_key
)

# Test call
response = llm.invoke("Say hello!")
print(response.content)
```

```python
import os, requests, datetime
key = os.getenv("OPENAI_API_KEY")
assert key, "Missing OPENAI_API_KEY"
headers = {"Authorization": f"Bearer {key}"}

# Credits
credits = requests.get("https://api.openai.com/v1/dashboard/billing/credit_grants", headers=headers).json()
print("Total granted:", credits.get("total_granted"))
print("Total used:", credits.get("total_used"))
print("Total available:", credits.get("total_available"))

# Last 30 days usage
end = datetime.date.today()
start = end - datetime.timedelta(days=30)
usage = requests.get(
    f"https://api.openai.com/v1/dashboard/billing/usage?start_date={start}&end_date={end}",
    headers=headers
).json()
print("Last 30 day usage (USD):", usage.get("total_usage")/100 if usage.get("total_usage") else None)
```
